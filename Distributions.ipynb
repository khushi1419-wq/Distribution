Q-3. A dataset has a mean of 50 and a standard deviation of 5. What percentage of values lie between 45 and
55 if the data follows a normal distribution?

## Answer:-

The values 45 and 55 are exactly 1 standard deviation away from the mean:

Mean = 50

Standard deviation = 5

45 = 50 − 5 (−1 SD)

55 = 50 + 5 (+1 SD)

Using the Empirical Rule (68–95–99.7 Rule):

About 68% of the data lies within ±1 standard deviation from the mean.

Approximately 68% of the values lie between 45 and 55.

## Interpretation:

This means most of the data is closely clustered around the mean, indicating moderate variability.

Q-4. What is the concept of standardization (z-score), and why is it important in data analysis? Explain the
formula and how standardization transforms a dataset.

## Answer:-

Concept of Standardization (Z-score)

Standardization, or the z-score, is a method used to convert data values into a common scale so they can be easily compared. It tells us how many standard deviations a value is away from the mean.

FORMULA

z= x- μ/σ

(z)≈88.5%

Where:

x = original value

μ (mu) = mean of the dataset

σ (sigma) = standard deviation

## How Standardization Transforms Data:

Shifts the data so that the new mean becomes 0

Rescales the data so that the standard deviation becomes 1

Converts all values into z-scores (positive, negative, or zero)

Positive z-score → above mean

Negative z-score → below mean

Zero → exactly at mean

## Why It Is Important in Data Analysis:

Makes data comparable → even if datasets have different units or scales

Helps detect outliers → extreme z-scores indicate unusual values

Used in statistical methods → like hypothesis testing and normal distribution analysis

Improves machine learning models → many algorithms work better with standardized data.

Q-5. What is Kurtosis and their type?

## Answer:-

Kurtosis is a statistical measure that describes the shape of a distribution, particularly the peakedness and heaviness of its tails (how extreme the values are compared to a normal distribution).

## Types of kurtosis:-

Mesokurtic:

→ Normal distribution

→ Moderate peak and tails

→ Example: Standard normal distribution

→ Kurtosis = 3

Leptokurtic:

→ Highly peaked distribution

→ Heavy tails (more outliers)

→ Data is concentrated near the mean but with extreme values

→ Kurtosis > 3

Platykurtic:

→ Flat distribution

→ Light tails (fewer outliers)

→ Data is more spread out

→ Kurtosis < 3

Q-6. Explain why the uniform distribution is a good model for the outcome of rolling a fair die.

## ANSWER:-

A uniform distribution is a probability distribution in which all outcomes are equally likely. This makes it a good model for rolling a fair die because:

A fair die has six possible outcomes (1, 2, 3, 4, 5, 6).

Each outcome has the same probability, i.e., 1/6.

There is no bias, so no number is more likely than another.

## Why it fits uniform distribution:

Equal probabilities → Every face has an equal chance.

Discrete outcomes → Limited and countable results.

Randomness → Each roll is independent and unbiased.

Q-8. How does log transformation help in stabilizing variance and making data more normally distributed?

## Answer:-

Log transformation means taking the logarithm (e.g., log₁₀ or ln) of each data value to reduce skewness and make patterns clearer.

## How it stabilizes variance:

In many datasets, variance increases with the mean (heteroscedasticity).

Log transformation compresses large values more than small ones.

This reduces the spread of high values and makes the variance more constant across the dataset.

## How it makes data more normal:

Right-skewed data (long tail on the right) becomes more symmetric after log transformation.

Extreme large values (outliers) are pulled closer to the center.

The distribution starts to resemble a normal (bell-shaped) curve.
